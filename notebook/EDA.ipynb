```python
df['Total_score']=df["math score"]+df["reading score"]+df["writing score"]
df["Avg"]=df["Total_score"]/3
df.head()
```

```python
math_full=df[df['math score']==100]['Avg'].count()
print(f"the number of students who scored full marks in maths are:{math_full}")

read_full=df[df['reading score']==100]['Avg'].count()
print(f"the number of students who scored full marks in reading  are:{read_full}")

write_full=df[df['writing score']==100]['Avg'].count()
print(f"the number of students who scored full marks  are:{write_full}")
```

```python
math_less=df[df['math score']<=20]['Avg'].count()
print(f"the number of students who scored less marks than 20 in maths are:{math_less}")

read_less=df[df['reading score']<=20]['Avg'].count()
print(f"the number of students who scored less marks than 20 in reading  are:{read_less}")

write_less=df[df['writing score']<=20]['Avg'].count()
print(f"the number of students who scored less  marks than 20  are:{write_less}")
```
